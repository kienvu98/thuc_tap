```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# đọc dữ liệu
A = np.genfromtxt("iris_full.csv", delimiter=",")
A
```

```python
y= A[:,-1]
X=A[:,0:4]
bias= np.ones((len(y),1))
X_bias= np.concatenate((bias, X), axis=1)
X_bias
```

```python
def sigmod(z):
    g = 1/(1 + np.exp(-z))
    return g

def loss(X,y,θ):
    z= X.dot(θ)
    g= sigmod(z)
    l= y.T.dot(np.log(g)) + (1 - y).T.dot(np.log(1 - g))
    return -l/len(y)

def gradient(X,y,θ):
    z= X.dot(θ)
    g= sigmod(z)
    t = g-y
    θ_gra = X.T.dot(t)
    return θ_gra/len(y)

def update(X,y,θ,n=0.01):
    θ_gra= gradient(X,y,θ)
    θ_new = θ - n*θ_gra
    return θ_new

cost=[]
θs=[]
θ= np.zeros(X_bias.shape[1])
for i in range(1000):
    t = loss(X_bias,y,θ)
    
    cost.append(t)
    θ = update(X_bias, y,θ)
    θs.append(θ)
    
cost
```

```python
t=[]
for i in range(len(cost)):
    t.append(i)
plt.plot(t,cost)
```
